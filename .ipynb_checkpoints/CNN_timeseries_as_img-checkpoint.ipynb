```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data_milk = pd.read_csv('monthly-milk-production-pounds-p.csv')[:-1:]
data_imports = pd.read_csv('monthly-australian-imports-from-.csv')[:-1:]
```

```python
data_milk_drop = data_milk.drop(['Month'], 1)
data_milk_drop.values.reshape(-1,1)
plt.plot(data_milk_drop)
plt.title('Milk production')
plt.show()
```

```python
data_imports_drop = data_imports.drop(['Month'], 1)
data_imports_drop.values.reshape(-1,1)
plt.plot(data_imports_drop)
plt.title('Australian imports')
plt.show()
```

```python
window_size = 16
pic_size = window_size*window_size
timeseries = np.array(data_milk_drop)
```

```python
def array_to_cell(timeseries, window_size):
    X = [list(itertools.chain.from_iterable(timeseries[start:start + window_size]))for start in range(0, timeseries.shape[0] - window_size)]
    label = [1.0 if timeseries[i] > timeseries[i-1] else 0.0  for i in range(window_size,len(timeseries))]
    return X, label
```

```python
def list_to_ing(x):
    x_image = np.zeros(shape = (x.shape[0], pic_size))
    for i in range(x.shape[0]):
        x_temp = x[i].copy()
        x_copy = x[i].copy()
        x_int = np.zeros(shape = (x_copy.shape), dtype = int)
        x_copy2 = x[i].copy()
        x_copy.sort()
        for j in range(x_copy2.size):
            for k in range(x_copy.size):
                   if x_copy2[j] == x_copy[k]:
                        x[i, j] = int(k)
        n_values = int(x[i].max())+1
        squared = np.eye(n_values)[x[i].astype(int)]
        flatten = squared.flatten()
        x_image[i, :] = flatten
    return x_image
    
```

```python
X, label = array_to_cell(timeseries, window_size)
x = np.copy(X)
```

```python
x_img = list_to_ing(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_img, label, test_size = 0.3) 
```

```python
def neural_net_model(X_data):
    input_layer_of_load_images = tf.reshape(x, [-1, 16, 16, 1])
    convolute_layer_1 = tf.layers.conv2d(inputs = input_layer_of_load_images, filters = 16, kernel_size = [5, 5], padding = "same", activation = tf.nn.relu)
    pool_layer_1 = tf.layers.max_pooling2d(inputs=convolute_layer_1, pool_size=[2, 2], strides=2)
    convolute_layer_2 = tf.layers.conv2d(inputs=pool_layer_1,filters=64,kernel_size=[3, 3],padding="same",activation=tf.nn.relu)
    pool_layer_2 = tf.layers.max_pooling2d(inputs = convolute_layer_2, pool_size=[2, 2],strides=2)
    
    pool_layer_2_flat = tf.reshape(pool_layer_2, [-1, 4*4*64])
    
    dense_layer_1 = tf.layers.dense(inputs = pool_layer_2_flat, units=1024,activation=tf.nn.relu)
    dense_layer_2 = tf.layers.dense(inputs = dense_layer_1, units = 400, activation=tf.nn.relu)
    dropout = tf.layers.dropout(inputs = dense_layer_2, rate = 0.3)
    
    dense_layer_3 = tf.layers.dense(inputs = dropout, units=200)
    prediction = tf.layers.dense(dense_layer_3, 2)
    return prediction
```

```python
x = tf.placeholder(tf.float32, [None, 256])   
y = tf.placeholder(tf.float32, [None, 2])
input_layer_of_load_images = tf.reshape(x, [-1, 16, 16, 1])  
output = neural_net_model(input_layer_of_load_images)
output_num = tf.argmax(output, 1)

#loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y, logits=output))
#train_op = tf.train.GradientDescentOptimizer(0.001).minimize(loss)

# calcualte average accuracy over a batch of images
#accuracy = tf.cast(tf.equal(tf.argmax(output, 1), tf.argmax(y, 1)), tf.float32)

cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y, logits=output))  
train_step = tf.train.AdamOptimizer(0.0001).minimize(cross_entropy)
accuracy = tf.cast(tf.equal(tf.argmax(output, 1), tf.argmax(y, 1)), tf.float32)
```
